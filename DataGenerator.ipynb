```python
start_date = datetime.date(2010, 1, 1)
end_date = datetime.date(2020, 2, 1)
for i in range(5000):
    hash.update(str(time.time()).encode('utf-8'))
    hash.hexdigest()[:10]
    time_between_dates = end_date - start_date
    days_between_dates = time_between_dates.days
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + datetime.timedelta(days=random_number_of_days)

    new_row = {'UserID':hash.hexdigest()[:10], 'UserEmail':str(hash.hexdigest()[:10]) + "@mail.com", 'TeamID':random.randint(1, 20), 'Name':names.get_full_name(), 'IsActive':random.randint(0, 1), 'DateRegister':random_date}
    #append row to the dataframe
    Users = Users.append(new_row, ignore_index=True)
Users
```

```python
for i in range(19):
    new_row = {'TeamID':i+1, 'TeamName':str(r.get_random_word()), 'TeamManager':names.get_full_name(), 'TeamPriority':random.randint(1, 5)}
    #append row to the dataframe
    Teams = Teams.append(new_row, ignore_index=True)
Teams
```

```python
end_date = datetime.date(2020, 2, 1)
for i in range(100000):
    sample = Users.sample()
    start_date = sample['DateRegister'].iloc[0]
    time_between_dates = end_date - start_date
    days_between_dates = time_between_dates.days
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + datetime.timedelta(days=random_number_of_days)

    new_row = {'UserId':sample['UserID'].iloc[0], 'DateMetric':random_date, 'Mails':random.randint(0, 60), 'Hits':random.randint(0, 50), 'SMS':random.randint(0, 35), 'PhoneCalls':random.randint(0, 20)}
    #append row to the dataframe
    Metrics = Metrics.append(new_row, ignore_index=True)
Metrics
```

```python
Metrics['DateMetric'] =  Metrics['DateMetric'].astype(str)
Users['DateRegister'] =  Users['DateRegister'].astype(str)
Metrics.to_parquet('Metrics.parquet', engine='fastparquet')
Teams.to_parquet('Teams.parquet', engine='fastparquet')
Users.to_parquet('Users.parquet', engine='fastparquet')
```
